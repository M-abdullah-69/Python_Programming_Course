```python
#1. Make a calculator using Python with addition , subtraction , multiplication ,division and power.

#class calculator
class Calculator:
    print("Welcome to the calculator: \n")
        
    #functions that perform different tasks
    def Add(self, x=0 , y=0 , *args):
        ans = x + y 
        if len(args) >= 1 :
            for i in args:
                ans += i
        return f"The addition of the numbers {x,y,*args} you given is {ans}"
    
    def Sub(self, x=0 , y=0 , *args):
        ans = x - y 
        if len(args) >= 1 :
            for i in args:
                ans -= i
        return f"The Subtraction of the numbers {x,y,*args} you given is {ans}"
    
    def Mul(self, x=0 , y=0 , *args):
        ans = x * y 
        if len(args) >= 1 :
            for i in args:
                ans *= i
        return f"The Multiplication of the numbers {x,y,*args} you given is {ans}"
    
    def Div(self, x=0 , y=0 , *args):
        ans = x // y 
        if len(args) >= 1 :
            for i in args:
                ans /= i
        return f"The Division of the numbers {x,y,*args} you given is {ans}"
    
    def Power(self, x=0 , y=0 , *args):
        ans = x ** y 
        if len(args) >= 1 :
            for i in args:
                ans **= i
        return f"The Power of the number {x,y,*args} you given is {ans}"
    

#Making object
num = Calculator()
print(num.Add(3,4))
print(num.Sub(4,4))
print(num.Mul(5,5))
print(num.Div(6,6))
print(num.Power(8,2))
print(num.Power(8,2,3) )       
```

```python

#2. Write a program to check if there is any numeric value in list using for loop.

def has_num(myList):
    if myList == []:
        return "Empty list can not be accepted as argument"
    new = []
    for value in myList:
        if isinstance(value, (int, float)):
            new.append(value)
    if len(new) > 0:
        return f"Success, list contains numeric values {new}"
    else:
        return "Not a numeric value exist."
has_num([1,2,3,4,"Jas"])
```

```python
# 3. Write a Python script to add a key to a dictionary.
class Dict:
    def __init__(self):
        self.dict = {}
        
    def add(self,item,value):
        self.dict[item] = value
        
    def update(self,item,value):
        if item in self.dict:
            self.dict[item] = value
        else:
            print("Kindly enter correct item")
    
    def delete(self,item,value):
        if item in self.dict.keys():
            if value == self.dict[item]:
                del self.dict[item]
        else:
            print("Kindly enter correct item")
            
x = Dict()
x.add(1,"Jansher")
x.add(2,"Ameen")
x.add(3,"Abdullah")
x.add(4,"Talha")
print(f"Dictionary = {x.dict}\n")
x.update(3,"Huzaifa")
print(f"updated Dictionary = {x.dict}\n")
x.delete(2,"Jansher")
print(f"After deleting an item from a dictionary = {x.dict}")
x.delete(2,"Ameen")
print("Oh! Sorry wrong value entered")
print(f"After deleting an item from a dictionary = {x.dict}")
```

```python
# 5. Write a program to identify duplicate values from list.

def Find_Duplicate(lst):
    new = []
    duplicate = []
    if len(lst) > 0:
        for i in lst:
            if not i in new:
                new.append(i)
            else:
                duplicate.append(i)
    print(f"""Complete List                                 {lst}
Duplicate value in the list                   {duplicate}
After removing duplicate values from the list {new}""")

lst = [1,2,3,4,5,6,4,3,2,3]
Find_Duplicate(lst)
```
